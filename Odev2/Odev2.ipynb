```python
while True:
    try:
        # 1, 2, 3: Kullanıcıdan süre bilgilerini al
        gunluk_uyku_suresi_girdi = input("Lütfen günlük uyku sürenizi (saat olarak) girin: ")
        gunluk_ders_suresi_girdi = input("Lütfen günlük ders sürenizi (saat olarak) girin: ")
        gunluk_serbest_zaman_suresi_girdi = input("Lütfen günlük serbest zaman sürenizi (saat olarak) girin: ")

        # 4: Girilen değerlerin sayı olup olmadığını kontrol et (float'a çevirmeye çalışarak)
        gunluk_uyku_suresi = float(gunluk_uyku_suresi_girdi)
        gunluk_ders_suresi = float(gunluk_ders_suresi_girdi)
        gunluk_serbest_zaman_suresi = float(gunluk_serbest_zaman_suresi_girdi)

        # 5: Girilen değerlerin negatif olmamasını sağla
        if gunluk_uyku_suresi < 0 or gunluk_ders_suresi < 0 or gunluk_serbest_zaman_suresi < 0:
            print("Geçersiz girdi! Süreler negatif olamaz. Lütfen tekrar girin.")
            continue

        # 6: Girilen sürelerin toplamı 24 saati geçiyorsa kullanıcıyı uyar ve girişleri tekrar al
        toplam_sure = gunluk_uyku_suresi + gunluk_ders_suresi + gunluk_serbest_zaman_suresi

        if toplam_sure > 24:
            print("Toplam süre 24 saati geçemez. Lütfen tekrar girin.")
            continue
        elif toplam_sure < 24: # Yeni eklenen kontrol
            eksik_sure = 24 - toplam_sure
            print(f"Eksik etkinlik girişi yaptınız: {eksik_sure} saat. Lütfen toplam 24 saate tamamlayın.")
            continue # Tekrar giriş iste

        # Tüm girdiler geçerliyse döngüden çık
        break

    except ValueError:
        print("Geçersiz girdi! Lütfen sayısal bir değer girin.")
    except KeyboardInterrupt:
        print("Kullanıcı tarafından durduruldu.")
        gunluk_uyku_suresi = 0 # Varsayılan değerler atanabilir
        gunluk_ders_suresi = 0
        gunluk_serbest_zaman_suresi = 0
        break

# 7: Uyku süresine göre uyku durumunu belirle
if gunluk_uyku_suresi < 6:
    uyku_durumu = "Yetersiz uyku"
elif 6 <= gunluk_uyku_suresi <= 8:
    uyku_durumu = "Normal uyku"
else:
    uyku_durumu = "Fazla uyku"

# 8: Ders süresine göre çalışma yoğunluğunu belirle
if gunluk_ders_suresi >= 6:
    ders_calisma_yogunlugu = "Yoğun çalışma günü"
else:
    ders_calisma_yogunlugu = "Normal çalışma günü"

# 9: Genel değerlendirme sonucunu ekrana yazdır
if gunluk_uyku_suresi < 6 and gunluk_ders_suresi >= 6:
    genel_degerlendirme = "Verimsiz gün"
else:
    genel_degerlendirme = "Dengeli gün"

print(f"\n--- Günlük Değerlendirme --- ")
print(f"Girilen Uyku Süresi: {gunluk_uyku_suresi} saat")
print(f"Girilen Ders Süresi: {gunluk_ders_suresi} saat")
print(f"Girilen Serbest Zaman Süresi: {gunluk_serbest_zaman_suresi} saat")
print(f"Uyku Durumu: {uyku_durumu}")
if gunluk_uyku_suresi < 8: # Yeni eklenen uyku notu
    print("Uyku notu: Az uyudunuz.")
print(f"Ders Çalışma Yoğunluğu: {ders_calisma_yogunlugu}")
print(f"Genel Değerlendirme: {genel_degerlendirme}")
```
